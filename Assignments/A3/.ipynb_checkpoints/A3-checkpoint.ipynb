```python
# Import Matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python
my_dict = {} # create empty dictionary
```

```python
sample = "This is a sample text. Sample text is a good example as It showcases how to count words in a text string."
# sample text to be parsed
```

```python
def make_dict(text):
    ''' make_dict: Make a dictionary from the imputted text
           - every unique word is a key
           - the value is the amount of times the
             key is present in text
    '''
    my_list = [] # create empty list
    my_list = text.split(' ') # make a list from the sample text (string). Every item is text seperated by spaces.
    for i in range(len(my_list)): # loop through the my_list
        my_list[i] = my_list[i].lower() # make every my_list item (sub-string) lowercase
        if '.' in my_list[i]: # if the current sub-string contains a '.'
            my_list[i] = my_list[i].replace('.','') # replace '.' with nothing
        if my_dict.get(my_list[i]): # if substring is in my_dict (dictionary)
            my_dict[my_list[i]] += 1 # add a one to the value
        else: # if substring is not a key in my_dict
            my_dict[my_list[i]] = 1 # initialize it with substring being key and value = 1
    return my_dict # return the created dictionary
```

```python
# run make_dict() (explained above) with same text (defined above)
# returned dictionary set to my_dict
my_dict = make_dict(sample)
```

```python
my_keys = list(my_dict.keys()) # make a list of all the keys in my_dict
```

```python
my_values = list(my_dict.values()) # make a list of all the values in my_dict
```

```python
# Code for generating bar chart
fig, ax = plt.subplots() # using matplotlib, create figure and axes subplots
ax.set_title('Number of Words from Input') # title of graph
ax.set_xlabel('Words') # x-axis label
ax.set_ylabel('Number of Times') # y-axis label
plt.xticks(rotation=90) # makes x-axis labels vertical so they can be read easier

# generate bar chart with specifications in argument
ax.bar(my_keys, my_values, width=1, color="orange", edgecolor="white", linewidth=.7) 

plt.show() # display bar chart
```
